```python
import pandas as pd
import matplotlib.pyplot as plot
import matplotlib.ticker as tick
import numpy as np
from datetime import datetime

df1 = pd.read_csv("covid_19_india.csv")

df1.dropna(axis = 0)

print(df1.head())


def ToFormattedNumber(x, y):
    if(x==0): return "0"
    if(x  / 1000 < 1000):
        return "{:0.2f}K".format(float(x/1000))
    elif(x / 1000000 < 1000):
        return "{:0.2f} Million".format((x/1000000))




df = df1[["Date", "Confirmed"]]
df = df.groupby("Date", as_index=False).sum()


def ToFormattedDate(x, y):
    date = datetime.strptime(df["Date"].values[x], "%Y-%m-%d")

    return datetime.strftime(date, "%b %Y")

#######################################################
#              Total Confirmed Cases over Time        #
#######################################################

plot.figure()
plot.plot(df["Date"].values, df["Confirmed"].values, color="purple")
plot.gca().get_yaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
plot.gca().get_xaxis().set_major_formatter(tick.FuncFormatter(ToFormattedDate))
plot.xticks(np.arange(0 , len(df["Date"].values), 90))
plot.xlabel("Date")
plot.ylabel("Confirmed Cases")
plot.title("Total Confirmed Cases - " + ToFormattedNumber(df["Confirmed"].values[-1].astype(int), 0))
plot.show()



#######################################################
#              Total Recoveries over Time             #
#######################################################

df = df1[["Date", "Cured"]]
df = df.groupby("Date", as_index=False).sum()  


plot.figure()
plot.plot(df["Date"].values, df["Cured"].values, color="green")
plot.gca().get_yaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
plot.gca().get_xaxis().set_major_formatter(tick.FuncFormatter(ToFormattedDate))
plot.xticks(np.arange(0 , len(df["Date"].values),  90))
plot.xlabel("Date")
plot.ylabel("Confirmed Recoveries")
plot.title("Total Cured - " + ToFormattedNumber(df["Cured"].values[-1].astype(int), 0))
plot.show()

        
#######################################################
#              Total Deaths over Time                 #
#######################################################    

df = df1[["Date", "Deaths"]]
df = df.groupby("Date", as_index=False).sum()  


plot.figure()
plot.plot(df["Date"].values, df["Deaths"].values, color="red")
plot.gca().get_yaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
plot.gca().get_xaxis().set_major_formatter(tick.FuncFormatter(ToFormattedDate))
plot.xticks(np.arange(0 , len(df["Date"].values),  90))
plot.xlabel("Date")
plot.ylabel("Confirmed deaths")
plot.title("Total Deaths - " + ToFormattedNumber(df["Deaths"].values[-1].astype(int), 0))
plot.show()



#######################################################
#              Daily Cases Vs Recoveries over time    #
#######################################################

df = df1[["Date", "Confirmed"]]
df = df.groupby("Date", as_index=False).sum()  

df_ = df.shift(1)

df["Confirmed"] = df["Confirmed"] - df_["Confirmed"]



df__ = df1[["Date", "Cured"]]
df__ = df__.groupby("Date", as_index=False).sum()  

df___ = df__.shift(1)

df__["Cured"] = df__["Cured"] - df___["Cured"]



plot.figure()
plot.plot(df["Date"].values, df["Confirmed"].values, color="blue")
plot.plot(df__["Date"].values, df__["Cured"].values, color="green")
plot.gca().get_yaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
plot.gca().get_xaxis().set_major_formatter(tick.FuncFormatter(ToFormattedDate))
plot.xticks(np.arange(0 , len(df["Date"].values),  90))
plot.xlabel("Date")
plot.ylabel("Cases")
plot.title("New Cases (Blue) Vs Recoveries(Green)")
plot.show()



#######################################################
#              Recovery Rate                          #
#######################################################



df = df1[["Cured", "Deaths"]]
df = df.sum()  

plot.figure()
plot.pie([df["Cured"] , df["Deaths"]], labels=["Recovered", "Deaths"], autopct='%1.2f%%')
plot.gca().get_yaxis().set_major_formatter(tick.ScalarFormatter())
plot.gca().get_xaxis().set_major_formatter(tick.ScalarFormatter())
plot.title("Recovery Rate - {:0.2f}%".format((df["Cured"] / (df["Cured"] + df["Deaths"]) * 100).item()) );
plot.show()






def ToFormattedNumber(x, y):
    if(x==0): return "0"
    if(x  / 1000 < 1000):
        return "{:0.2f}K".format(float(x/1000))
    elif(x / 1000000 < 1000):
        return "{:0.2f} M".format((x/1000000))


    
    
#######################################################
#              Total Cases Per State                  #
#######################################################
    
df = df1[["State/UnionTerritory", "Confirmed"]]
df = df.groupby("State/UnionTerritory", as_index=False).last()
df= df.sort_values(["Confirmed"], axis=0)
plot.figure(figsize=(8, 12))
plot.barh(df["State/UnionTerritory"].values, df["Confirmed"].values)
# plot.gca().get_yaxis().set_major_formatter(tick.ScalarFormatter(ToFormattedNumber))
plot.gca().get_xaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
plot.title("Total Cases per State (In Descending Order)");

for i,v in enumerate(df["Confirmed"].values):
    plot.text(v, i, " "+ ToFormattedNumber(v.item(),0), color='blue', va='center')

#plot.xticks(np.arange(0 , max(df["Confirmed"].values), 1000000))
plot.show()



#######################################################
#               Recoveries Per State                  #
#######################################################


df = df1[["State/UnionTerritory", "Cured"]]
df = df.groupby("State/UnionTerritory", as_index=False).last()
df= df.sort_values(["Cured"], axis=0)
plot.figure(figsize=(8, 12))
plot.barh(df["State/UnionTerritory"].values, df["Cured"].values, color="green")
# plot.gca().get_yaxis().set_major_formatter(tick.ScalarFormatter(ToFormattedNumber))
plot.gca().get_xaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
plot.title("Total Recoveries per State (In Descending Order)");

for i,v in enumerate(df["Cured"].values):
    plot.text(v, i, " "+ ToFormattedNumber(v.item(),0), color='blue', va='center')

#plot.xticks(np.arange(0 , max(df["Confirmed"].values), 1000000))
plot.show()


#######################################################
#              Deaths per State                       #
#######################################################


df = df1[["State/UnionTerritory", "Deaths"]]
df = df.groupby("State/UnionTerritory", as_index=False).last()
df= df.sort_values(["Deaths"], axis=0)
plot.figure(figsize=(8, 12))
plot.barh(df["State/UnionTerritory"].values, df["Deaths"].values, color="red")
# plot.gca().get_yaxis().set_major_formatter(tick.ScalarFormatter(ToFormattedNumber))
plot.gca().get_xaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
plot.title("Total Deaths per State (In Descending Order)");

for i,v in enumerate(df["Deaths"].values):
    plot.text(v, i, " "+ ToFormattedNumber(v.item(),0), color='blue', va='center')

#plot.xticks(np.arange(0 , max(df["Confirmed"].values), 1000000))
plot.show()



#######################################################
#              Recovery Rate in % per state           #
#######################################################


df = df1[["State/UnionTerritory", "Deaths", "Cured"]]
df = df.groupby("State/UnionTerritory", as_index=False).last()

df["RecoveryRate"] = df["Cured"] / (df["Deaths"] + df["Cured"] )  * 100

df = df.dropna()
df= df.sort_values(["RecoveryRate"], axis=0)


plot.figure(figsize=(8, 12))
plot.barh(df["State/UnionTerritory"].values, df["RecoveryRate"].values, color="green")
plot.gca().get_xaxis().set_major_formatter(tick.PercentFormatter(100))
plot.title("Recovery Percentage");


for i,v in enumerate(df["RecoveryRate"].values):
    plot.text(v, i, " {:.2f}%".format(v.item() ), color='blue', va='center')

#plot.xticks(np.arange(0 , max(df["Confirmed"].values), 1000000))
plot.show()



df1 = pd.read_csv("StatewiseTestingDetails.csv")
#######################################################
#              Daily tests In India                   #
#######################################################




# df = df1[["State", "Date", "TotalSamples"]].groupby("Date", as_index=False).sum()
# df_ = df.shift(1)
# df["DailySamples"] = df["TotalSamples"] - df_["TotalSamples"]

# print(df.head())

# def ToFormattedDate(x, y):
#     date = datetime.strptime(df["Date"].values[x], "%Y-%m-%d")

#     return datetime.strftime(date, "%b %Y")


# plot.figure(figsize=(8, 12))
# plot.plot(df["Date"], df["TotalSamples"] , color = "yellow")
# plot.xlabel("Samples Per Day")
# plot.ylabel("Date")
# #plot.gca().get_yaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
# #plot.gca().get_xaxis().set_major_formatter(tick.FuncFormatter(ToFormattedDate))
# plot.title("Daily Tests in India")
# plot.show()


#######################################################
#              Total Tests Per State                  #
#######################################################



df = df1[["State", "TotalSamples", "Positive"]]
df_ = df1[["State", "Positive"]].groupby("State" , as_index=False).sum()
df = df.groupby("State", as_index=False).last()

print(df1.head(100))

plot.figure(figsize=(8, 12))
plot.barh(df["State"].values, df["TotalSamples"].values, color="green")
#plot.gca().get_xaxis().set_major_formatter(tick.PercentFormatter(100))
for i,v in enumerate(df["TotalSamples"].values):
    plot.text(v, i, " "+ ToFormattedNumber(v.item(),0), color='blue', va='center')


plot.gca().get_xaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
plot.title("Total Tests per State (In Descending Order)");

plot.show()


#######################################################
#              Vaccine Doses Administered             #
#######################################################


df1 = pd.read_csv("covid_vaccine_statewise.csv")

print(df1.head())

df = df1[df1["State"] == "India"]
#df = df.groupby("Updated On", as_index=False)

def ToFormattedDate(x, y):
    date = datetime.strptime(df["Updated On"].values[x], "%d/%m/%Y")

    return datetime.strftime(date, "%b %Y")

df.dropna()



plot.figure()
plot.plot(df["Updated On"].values, df["Total Doses Administered"].values, color="purple", label="Total")
plot.plot(df["Updated On"].values, df["Total CoviShield Administered"].values, color="cyan", label="Covi-Shield")
plot.plot(df["Updated On"].values, df["Total Covaxin Administered"].values, color="green", label="Covaxin")
plot.plot(df["Updated On"].values, df["Total Sputnik V Administered"].values, color="pink", label="Sputnik")

plot.legend()

plot.gca().get_yaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
plot.gca().get_xaxis().set_major_formatter(tick.FuncFormatter(ToFormattedDate))
plot.xticks(np.arange(0 , len(df["Updated On"].values), 30) )
plot.xlabel("Date")
plot.ylabel("Doses")
plot.title("Total Doses Administered")
plot.show()

#######################################################
#              Vaccine Market Share                   #
#######################################################



df = df1[df1["State"] == "India"]



plot.figure()
plot.pie([df["Total CoviShield Administered"].sum(), df["Total Covaxin Administered"].sum(), df["Total Sputnik V Administered"].sum()], labels=["Covi-Shield", "Covaxin", "Sputnik"], autopct='%1.5f%%')
plot.gca().get_yaxis().set_major_formatter(tick.ScalarFormatter())
plot.gca().get_xaxis().set_major_formatter(tick.ScalarFormatter())

plot.title("Market Share Of Each Vaccine")
plot.show()


#######################################################
#              First vs Second Dose                   #
#######################################################


df = df1[df1["State"] == "India"]
df = df.dropna()
df=df.iloc[-1]

plot.figure()
plot.barh(["First Dose", "Second Dose" ], [df["First Dose Administered"], df["Second Dose Administered"]],  label="Administered Doses",  color="purple")
# plot.gca().get_yaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
plot.gca().get_xaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
# plot.xticks(np.arange(0 , len(df["Date"].values), 90))
plot.xlabel("Doses")
plot.ylabel("Dose Number")
plot.title("Each Dose Count")
plot.show()



#######################################################
#              Doses By Age Group                     #
#######################################################



df = df1[df1["State"] == "India"]
#df = df.groupby("Updated On", as_index=False)

def ToFormattedDate(x, y):
    date = datetime.strptime(df["Updated On"].values[x], "%d/%m/%Y")

    return datetime.strftime(date, "%b %Y")

df = df.dropna(subset=["18-45 years (Age)", "45-60 years (Age)", "60+ years (Age)"])


plot.figure()
plot.plot(df["Updated On"].values, df["18-45 years (Age)"].values, color="purple", label="18-45 yr")
plot.plot(df["Updated On"].values, df["45-60 years (Age)"].values, color="cyan", label="45-60 yr")
plot.plot(df["Updated On"].values, df["60+ years (Age)"].values, color="green", label="60+ yr")
# plot.plot(df["Updated On"].values, df["Total Sputnik V Administered"].values, color="pink", label="Sputnik")

plot.legend()

plot.gca().get_yaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
plot.gca().get_xaxis().set_major_formatter(tick.FuncFormatter(ToFormattedDate))
plot.xticks(np.arange(0 , len(df["Updated On"].values), 30) )
plot.xlabel("Date")
plot.ylabel("Doses")
plot.title("Doses By Age-Group")
plot.show()



#######################################################
#           Doses Administratored By Each State       #
#######################################################


df = df1[df1["State"] != "India"]

df = df[["State", "Total Doses Administered"]]
df = df.groupby("State", as_index=False).last()
df= df.sort_values(["Total Doses Administered"], axis=0)
plot.figure(figsize=(8, 12))
plot.barh(df["State"].values, df["Total Doses Administered"].values, color="magenta")
# plot.gca().get_yaxis().set_major_formatter(tick.ScalarFormatter(ToFormattedNumber))
plot.gca().get_xaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
plot.title("Total Doses Administered per State");

for i,v in enumerate(df["Total Doses Administered"].values):
    plot.text(v, i, " "+ ToFormattedNumber(v.item(),0), color='blue', va='center')

#plot.xticks(np.arange(0 , max(df["Confirmed"].values), 1000000))
plot.show()

```
