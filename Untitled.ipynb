```python
import pandas as pd
import matplotlib.pyplot as plot
import matplotlib.ticker as tick
import numpy as np
from datetime import datetime

df1 = pd.read_csv("covid_19_india.csv")

print(df1.head())


def ToFormattedNumber(x, y):
    if(x==0): return "0"
    if(x  / 1000 < 1000):
        return "{:0.2f}K".format(float(x/1000))
    elif(x / 1000000 < 1000):
        return "{:0.2f} Million".format((x/1000000))




df = df1[["Date", "Confirmed"]]
df = df.groupby("Date", as_index=False).sum()


def ToFormattedDate(x, y):
    date = datetime.strptime(df["Date"].values[x], "%Y-%m-%d")

    return datetime.strftime(date, "%b %Y")


plot.figure()
plot.plot(df["Date"].values, df["Confirmed"].values, color="purple")
plot.gca().get_yaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
plot.gca().get_xaxis().set_major_formatter(tick.FuncFormatter(ToFormattedDate))
plot.xticks(np.arange(0 , len(df["Date"].values), 90))
plot.xlabel("Date")
plot.ylabel("Confirmed Cases")
plot.title("Total Confirmed Cases - " + ToFormattedNumber(df["Confirmed"].values[-1].astype(int), 0))
plot.show()
    

df = df1[["Date", "Cured"]]
df = df.groupby("Date", as_index=False).sum()  


plot.figure()
plot.plot(df["Date"].values, df["Cured"].values, color="green")
plot.gca().get_yaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
plot.gca().get_xaxis().set_major_formatter(tick.FuncFormatter(ToFormattedDate))
plot.xticks(np.arange(0 , len(df["Date"].values),  90))
plot.xlabel("Date")
plot.ylabel("Confirmed Recoveries")
plot.title("Total Cured - " + ToFormattedNumber(df["Cured"].values[-1].astype(int), 0))
plot.show()
        
    

df = df1[["Date", "Deaths"]]
df = df.groupby("Date", as_index=False).sum()  


plot.figure()
plot.plot(df["Date"].values, df["Deaths"].values, color="red")
plot.gca().get_yaxis().set_major_formatter(tick.FuncFormatter(ToFormattedNumber))
plot.gca().get_xaxis().set_major_formatter(tick.FuncFormatter(ToFormattedDate))
plot.xticks(np.arange(0 , len(df["Date"].values),  90))
plot.xlabel("Date")
plot.ylabel("Confirmed deaths")
plot.title("Total Deaths - " + ToFormattedNumber(df["Deaths"].values[-1].astype(int), 0))
plot.show()




df = df1[["Cured", "Deaths"]]
df = df.sum()  

plot.figure()
plot.pie([df["Cured"] , df["Deaths"]], labels=["Recovered", "Deaths"], autopct='%1.2f%%')
plot.gca().get_yaxis().set_major_formatter(tick.ScalarFormatter(ToFormattedNumber))
plot.gca().get_xaxis().set_major_formatter(tick.ScalarFormatter(ToFormattedDate))
plot.title("Recovery Rate - {:0.2f}%".format((df["Cured"] / (df["Cured"] + df["Deaths"]) * 100).item()) );
plot.show()


```
